```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
infile1='data/PREM.dat'
rprem    = np.loadtxt(fname=infile1,skiprows=4,usecols=0)
rhoprem  = np.loadtxt(fname=infile1,skiprows=4,usecols=3)
gzprem   = np.loadtxt(fname=infile1,skiprows=4,usecols=4)
```

----
## Constant density (outside Earth)

The gravitational acceleration **outside** of the Earth is given as:
$$
|g_{out}(r)| = \frac{GM}{r^2}, \quad r > R
$$
with
- Grav. constant $G=6.6741 \times 10^{-11}$ [m$^3$/kg/s$^2$]
- Earth mass $M=5.97 \times 10^{24}$ kg
- Earth radius$R=6371$ km

```python
ðº = 6.67408e-11
M = 5.972e24
rhoa = 5511.

rout = np.arange(6371,20000,50)
gzout = G*M/(rout*1000)**2

fig,ax1 = plt.subplots(figsize=(8,8))
ax1.set_xlabel('Radius [km]')
ax1.set_xlim([0,12000])
ax1.set_ylim([0,12])
ax1.set_ylabel('Gravity [m/s$^2$]')
ax1.plot([6371,6371],[0,14],linewidth='6',linestyle='-',color='black')
ax1.plot(rout,gzout,linewidth='2',color='red',label='outside $g$')
ax1.legend()

ax2 = ax1.twinx()
ax2.set_ylim([0,14])
ax2.set_ylabel('Density [10$^3$ kg/m$^3$]')
ax2.legend(loc='lower right')
```

## Constant density

The gravitational acceleration **inside** of the Earth for constant density $\rho$ is given as:
$$
|g_{in}(r)| = \frac{4}{3} \pi G \rho r, \quad r \le R
$$
with
- Grav. constant $G=6.6741 \times 10^{-11}$ [m$^3$/kg/s$^2$]
- Earth mass $M=5.97 \times 10^{24}$ kg
- Earth radius$R=6371$ km

```python
ðº = 6.67408e-11
M = 5.972e24
rhoa = 5511.

rin  = np.arange(0,6372,1)
rout = np.arange(6371,20000,50)
gzin = 0.*rin
rho  = 0.*rin
for i in range(len(rin)):
    rho[i]  = rhoa
    gzin[i] = 4./3.*np.pi*G*rho[i]*(rin[i]*1000)
gzout = G*M/(rout*1000)**2

fig,ax1 = plt.subplots(figsize=(8,8))
ax1.set_xlabel('Radius [km]')
ax1.set_xlim([0,12000])
ax1.set_ylim([0,12])
ax1.set_ylabel('Gravity [m/s$^2$]')
ax1.plot([6371,6371],[0,14],linewidth='6',linestyle='-',color='black')
ax1.plot(rprem,gzprem,linewidth='5',color=(0.8,0.8,0.8),label='PREM $g$')

ax1.plot(rin,gzin,linewidth='3',color='red',label='inside $g$')
ax1.plot(rout,gzout,linewidth='2',color='red',label='outside $g$')

ax1.legend()
ax2 = ax1.twinx()
ax2.set_ylim([0,14])
ax2.set_ylabel('Density [10$^3$ kg/m$^3$]')
ax2.plot(rprem,rhoprem/1000,linewidth='5',linestyle='--',color=(0.8,0.8,0.8),label='PREM $\\rho$')
ax2.plot(rin,rho/1000,linewidth='2',linestyle='--',color='red',label='$\\rho$')
ax2.legend(loc='lower right')
```

## Linear density distribution

$$
\begin{array}{rcl}
g &=& \frac{G M}{R^2} \\
&=& \frac{G}{R^2} \int_V \rho(r) dV \\
&=& \frac{G}{R^2} \int_V \rho(r) r^2 \sin\Theta d\Theta d\Phi dr \\
&=& \frac{4\pi G}{R^2} \int_0^R \rho(r) r^2 dr
\end{array}
$$
One shell:
$$
\begin{array}{rcl}
g &=& \frac{4\pi G}{R^2} \int_0^R \rho(r) r^2 dr \\
&=& \frac{4\pi G}{R^2} \rho \frac{R^3}{3} \\
&=& \frac{4}{3} \pi G \rho R
\end{array}
$$

```python
ðº = 6.67408e-11
M = 5.972e24
rho0 = 13000.
rho1 = 5500. #3300. #5500.

rin  = np.arange(0,6371,1)
rout = np.arange(6371,20000,50)
gzin = 0.*rin
rho  = 0.*rin
for i in range(len(rin)):
    rho[i]  = rho0 - (rho0-rho1)*rin[i]/6371
    gzin[i] = 4./3.*np.pi*G*rho[i]*(rin[i]*1000)
gzout = G*M/(rout*1000)**2

fig,ax1 = plt.subplots(figsize=(8,8))
ax1.set_xlabel('Radius [km]')
ax1.set_xlim([0,12000])
ax1.set_ylim([0,12])
ax1.set_ylabel('Gravity [m/s$^2$]')
ax1.plot([6371,6371],[0,14],linewidth='6',linestyle='-',color='black')
ax1.plot(rprem,gzprem,linewidth='5',color=(0.8,0.8,0.8),label='PREM $g$')

ax1.plot(rin,gzin,linewidth='3',color='red',label='inside $g$')
ax1.plot(rout,gzout,linewidth='2',color='red',label='outside $g$')

ax1.legend()
ax2 = ax1.twinx()
ax2.set_ylim([0,14])
ax2.set_ylabel('Density [10$^3$ kg/m$^3$]')
ax2.plot(rprem,rhoprem/1000,linewidth='5',linestyle='--',color=(0.8,0.8,0.8),label='PREM $\\rho$')
ax2.plot(rin,rho/1000,linewidth='2',linestyle='--',color='red',label='$\\rho$')
ax2.legend(loc='lower right')
```

```python
ðº = 6.67408e-11
M = 5.972e24
rho0 = 13000.
rho1 = 3000.

rin  = np.arange(0,6371,1)
rout = np.arange(6371,20000,50)
gzin = 0.*rin
rho  = 0.*rin
mass = 0.
for i in range(1,len(rin)):
    rho[i]  = rho0 - (rho0-rho1)*rin[i]/6371
    mass = mass + rho[i]*((rin[i]*1000)**3.-(rin[i-1]*1000)**3.) 
    gzin[i] = 4./3.*np.pi*G/(rin[i]*1000)**2.*mass
gzout = G*M/(rout*1000)**2

fig,ax1 = plt.subplots(figsize=(8,8))
ax1.set_xlabel('Radius [km]')
ax1.set_xlim([0,12000])
ax1.set_ylim([0,12])
ax1.set_ylabel('Gravity [m/s$^2$]')
ax1.plot([6371,6371],[0,14],linewidth='6',linestyle='-',color='black')
ax1.plot(rprem,gzprem,linewidth='5',color=(0.8,0.8,0.8),label='PREM $g$')

ax1.plot(rin,gzin,linewidth='3',color='red',label='inside $g$')
ax1.plot(rout,gzout,linewidth='2',color='red',label='outside $g$')

ax1.legend()
ax2 = ax1.twinx()
ax2.set_ylim([0,14])
ax2.set_ylabel('Density [10$^3$ kg/m$^3$]')
ax2.plot(rprem,rhoprem/1000,linewidth='5',linestyle='--',color=(0.8,0.8,0.8),label='PREM $\\rho$')
ax2.plot(rin,rho/1000,linewidth='2',linestyle='--',color='red',label='$\\rho$')
ax2.legend(loc='lower right')
```

## Density from PREM

$n$ shells:
$$
\begin{array}{rcl}
g &=& \frac{4\pi G}{R^2} \int_0^R \rho(r) r^2 dr \\ 
&=& \frac{4\pi G}{3 R^2} \left[ \sum\limits_{i=2}^{n} \rho_i (r_{i}^3-r_{i-1}^3) \right]
\end{array}
$$

```python
ðº = 6.67408e-11
M = 5.972e24
rho0 = 13000.
rho1 = 5500.

rin  = np.arange(0,6371,1)
rout = np.arange(6371,20000,50)
gzin = 0.*rin
rho  = 0.*rin
mass = 0.
for i in range(1,len(rin)):
    rho[i]  = rhoprem[i]
    mass = mass + rhoprem[i]*((rin[i]*1000)**3.-(rin[i-1]*1000)**3.)
    gzin[i] = 4./3.*np.pi*G/(rin[i]*1000)**2.*mass
gzout = G*M/(rout*1000)**2

fig,ax1 = plt.subplots(figsize=(8,8))
ax1.set_xlabel('Radius [km]')
ax1.set_xlim([0,12000])
ax1.set_ylim([0,12])
ax1.set_ylabel('Gravity [m/s$^2$]')
ax1.plot([6371,6371],[0,14],linewidth='6',linestyle='-',color='black')
ax1.plot(rprem,gzprem,linewidth='5',color=(0.8,0.8,0.8),label='PREM $g$')

ax1.plot(rin,gzin,linewidth='3',color='red',label='inside $g$')
ax1.plot(rout,gzout,linewidth='2',color='red',label='outside $g$')

ax1.legend()
ax2 = ax1.twinx()
ax2.set_ylim([0,14])
ax2.set_ylabel('Density [10$^3$ kg/m$^3$]')
ax2.plot(rprem,rhoprem/1000,linewidth='5',linestyle='--',color=(0.8,0.8,0.8),label='PREM $\\rho$')
ax2.plot(rin,rho/1000,linewidth='2',linestyle='--',color='red',label='$\\rho$')
ax2.legend(loc='lower right')
```

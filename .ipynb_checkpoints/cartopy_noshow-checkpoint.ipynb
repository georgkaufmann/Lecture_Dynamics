```python
import matplotlib.pyplot as plt
import numpy as np

import cartopy.crs as ccrs
import cartopy.feature as cfeature
import cartopy.mpl.ticker as cticker
```

----
## Using `Projections`

Global map in Robinson projection, with coastlines added.

```python
fig = plt.figure(figsize=(10,5))
ax = plt.axes(projection=ccrs.Robinson())
ax.coastlines(color='blue')
plt.title("Robinson Projection")
```

----
## Using `gridlines`

Add gridlines to map.

```python
fig = plt.figure(figsize=(10,5))
ax = plt.axes(projection=ccrs.Robinson())
ax.coastlines(color='blue')
ax.gridlines(draw_labels=True, dms=True, x_inline=False, y_inline=False,color='black')
plt.title("Robinson Projection")
```

----
## Using `features`

Add a land-ocean mask, draw coastlines, and add lakes and rivers.

```python
fig = plt.figure(figsize=(10,5))
ax = plt.axes(projection=ccrs.Robinson())
ax.add_feature(cfeature.LAND,color='green')
ax.add_feature(cfeature.OCEAN)
ax.add_feature(cfeature.COASTLINE,color='black')
ax.add_feature(cfeature.LAKES, alpha=0.5)
ax.add_feature(cfeature.RIVERS,color='blue')
ax.gridlines(draw_labels=True, dms=True, x_inline=False, y_inline=False,color='black')
plt.title("Robinson Projection")
```

----
## Change `extent`

```python
fig = plt.figure(figsize=(10,5))
ax = plt.axes(projection=ccrs.PlateCarree())
ax.set_extent([-10, 10, 45, 60])
ax.gridlines(draw_labels=True, dms=True, x_inline=False, y_inline=False,color='black')
ax.add_feature(cfeature.OCEAN)
ax.add_feature(cfeature.COASTLINE,color='black')
ax.add_feature(cfeature.LAKES, alpha=0.5)
ax.add_feature(cfeature.RIVERS,color='blue')
plt.title("PlateCarree Projection")
```

----
## Add custom labels

```python
fig = plt.figure(figsize=(10,5))
ax = plt.axes(projection=ccrs.PlateCarree())
ax.set_extent([-10, 10, 45, 60])
ax.set_xticks([-8,-6,-4,-2,0,2,4,6,8])
ax.set_yticks([46,48,50,52,54,56,58])
lon_formatter = cticker.LongitudeFormatter(number_format='.1f',zero_direction_label=True)
lat_formatter = cticker.LatitudeFormatter(number_format='.1f')
ax.xaxis.set_major_formatter(lon_formatter)
ax.yaxis.set_major_formatter(lat_formatter)
ax.gridlines(draw_labels=False,linewidth=1, color='gray', alpha=0.5, linestyle='--')
ax.add_feature(cfeature.OCEAN)
ax.add_feature(cfeature.COASTLINE,color='black')
ax.add_feature(cfeature.LAKES, alpha=0.5)
ax.add_feature(cfeature.RIVERS,color='blue')
plt.title("PlateCarree Projection")
```

Use the standard image in `cartopy`.

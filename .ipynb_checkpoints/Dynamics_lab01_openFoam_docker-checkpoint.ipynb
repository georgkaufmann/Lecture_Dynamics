## Installing and running `docker`

From wikipedia: **Docker is a computer program that performs operating-system-level 
virtualization also known as containerization. It is developed by Docker, Inc.**

### Install docker

1. Download `docker` from [www.docker.com](http://www.docker.com) for your
operating system. Install `docker` and try to start the `docker` daemon.
The `docker` daemon runs in the background. Test its functionality by opening
a command-line shell (Linux: e.g. xterm,console; Windows: DOs prompt) and
type 'docker --version'. If `docker` is running, the version number should
be returned:

~~~bash
# docker --version
Docker version 19.03.5, build 633a0ea
~~~

2. Try `docker --help` for a list of commands.

### Downloading a repository

Once `docker` is running on your computer, you can start downloading *repositories*
from the docker hub, e.g. the `ubuntu` image with the command:

~~~bash
# docker pull ubuntu:20.04
~~~

### Running ubuntu in docker

Check, if the docker image is downloaded to your computer:

~~~bash
# docker images

REPOSITORY                          TAG                 IMAGE ID            CREATED             SIZE
...
ubuntu                              20.04               1e4467b07108        14 months ago       73.9MB
openfoam/openfoam8-graphical-apps   latest              c2e1cd53823d        7 weeks ago         1.76GB
...
~~~

The result should list all accessible images.

Next, try to run the `ubuntu` image with the command:

~~~bash
$ docker run -it --rm  ubuntu:20.04
~~~

Now a so-called instance of `docker` should start, providing a virtual linux machine, offering
a command prompt ($). Try a few linux commands in the virtual container:

~~~bash
$ pwd
/
$ ls
bin  boot  dev  etc  home  lib  lib32  lib64  libx32  media  mnt  opt  proc  root  run  sbin  srv  sys  tmp  usr  var
$ cd home
home# pwd
/home

$ cat /etc/os-release
NAME="Ubuntu"
VERSION="20.04 LTS (Focal Fossa)"
ID=ubuntu
ID_LIKE=debian
PRETTY_NAME="Ubuntu 20.04 LTS"
VERSION_ID="20.04"
HOME_URL="https://www.ubuntu.com/"

BUG_REPORT_URL="https://bugs.launchpad.net/ubuntu/"
PRIVACY_POLICY_URL="https://www.ubuntu.com/legal/terms-and-policies/privacy-policy"
VERSION_CODENAME=focal
UBUNTU_CODENAME=focal
~~~

To leave the docker container environment, type `exit` at the command prompt.

~~~bash
$ exit
~~~

## Running openFoam in docker

Next, try to run this repository with the command:

~~~bash
$ docker run -it --rm -v $PWD/tmp:/home/tmp openfoam/openfoam9-graphical-apps
~~~

This command starts a `docker container`, and runs an interactive shell.

It also links the local directory `tmp` to the directory `/home/tmp`  in the
docker container. This directory we wil use to make file exchange between
local computer and docker container easy.

Check the container with:

~~~bash
$ docker container ls

CONTAINER ID        IMAGE                   COMMAND                  CREATED             STATUS              PORTS                  NAMES
5faf5151f3df        ubuntu:20.04            "/bin/bash"              6 seconds ago       Up 6 seconds                               amazing_turing
7a4b9aa2456d        openfoam-v2106          "/bin/bash"              3 hours ago         Up 3 hours
~~~

### General variables

`openFoam` has several shell variables, which it uses, e.q.

~~~bash
$ echo $FOAM_RUN
/home/openfoam/run
$FOAM_TUTORIALS
/opt/openfoam8/tutorials
~~~

From the latter path, you can copy the tutorials.

If the shell variables are not set, then it is likely that a `bashrc` script has not been
executed, likely the one in `$FOAM_RUN/etc/bashrc`. Run this script with the
command `source` (path needs to be adopted):

~~~bash
$ source /opt/openfoam/etc/bashrc
~~~

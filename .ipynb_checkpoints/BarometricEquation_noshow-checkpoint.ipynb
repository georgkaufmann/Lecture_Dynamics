## Luftdruck

[Barometric pressure](https://de.wikipedia.org/wiki/Barometrische_H%C3%B6henformel)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
TC = 15.
T0 = 273.15 + TC # C
p0 = 1013.25     # hPa
dT0 = 0.0065     # K/m 
```

```python
h = np.linspace(0,11000,101)
p = p0 * (1 - dT0*h/T0)**5.255
```

```python
plt.figure(figsize=(12,6))
plt.ylabel('Pressure [hPa]')
plt.xlabel('Elevation [m]')
plt.grid()
plt.plot(h,p,linewidth=2,label='p')
```

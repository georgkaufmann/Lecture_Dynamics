An initial amount $A$ [€] will reduce to $B$ within time $t$ [a], if the interest rate is $p$:
$$
B = A \left( 1 - p \right)^t
$$

For a given time $t$, we then need the interest rate:
$$
p = 1 - \sqrt[t]{ \frac{B}{A}}
$$

```python
AList = np.array([11e6,20e6,100e6,1e9,10e9,100e9])
pList = np.zeros_like(AList)
B = 10e6
t = 10.

for i,A in enumerate(AList):
    pList[i] = 1 - (B/A)**(1/t)
    BB = A*(1-pList[i])**t
    #print(i,A,pList[i],BB)
```

```python
plt.xlim([0.9*AList.min()/1e6,1.1*AList.max()/1e6])
plt.xlabel('Amount [Mill €]')
plt.ylim([-5,80])
plt.ylabel('Interest rate [%]')
plt.plot(AList/1e6,pList*100,lw=4,marker='o',color='red',markersize=12,markerfacecolor='blue',markeredgecolor='black')
for i in range(AList.shape[0]):
    plt.plot([10,AList[i]/1e6],[pList[i]*100,pList[i]*100],lw=2,ls='dashed',color='blue',alpha=0.5)
plt.xscale('log')
plt.grid()
```

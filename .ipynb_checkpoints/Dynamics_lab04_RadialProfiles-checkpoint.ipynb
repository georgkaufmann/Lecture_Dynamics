In this notebook, we read in **PREM** data, plot mechanical parameters such as
- bulk-wave velocities
- shear-wave velocities
- bulk and shear modulus
- Poisson ratio
- Density
- Gravitational acceleration

and finally we calculate
- geotherms

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.transforms as transforms
import scipy.special
```

----
## Read PREM data

```python
infile1='data/PREM.dat'
radius   = np.loadtxt(fname=infile1,skiprows=4,usecols=0)
vp       = np.loadtxt(fname=infile1,skiprows=4,usecols=1)
vs       = np.loadtxt(fname=infile1,skiprows=4,usecols=2)
density  = np.loadtxt(fname=infile1,skiprows=4,usecols=3)
gravity  = np.loadtxt(fname=infile1,skiprows=4,usecols=4)
poisson  = np.loadtxt(fname=infile1,skiprows=4,usecols=5)
shear    = np.loadtxt(fname=infile1,skiprows=4,usecols=8)
bulk     = np.loadtxt(fname=infile1,skiprows=4,usecols=10)
```

----
## Calculate lithostatic pressure

$$
p = \int \rho g dz
$$

```python
pressure = np.zeros(len(radius))
for i in range(len(radius)-2,-1,-1):
    pressure[i] = pressure[i+1] + density[i]*gravity[i]*(radius[i+1]-radius[i])*1000
```

----
## Plot radial dependencies ...

```python
def boundaries(ax):
    """
    function defines boundaries in the Earth and labels them
    We use the data range (here [0,6371]) in y direction,
    but the graphics range [0,1] in x direction
    """
    trans = transforms.blended_transform_factory(ax.transAxes, ax.transData)
    ax.plot([0,1],[6371-440,6371-440],transform=trans,linewidth=6,color="lightgray")
    ax.text(0.1,6371-440,"440 km",transform=trans,fontweight="bold",verticalalignment='center')
    ax.plot([0,1],[6371-660,6371-660],transform=trans,linewidth=6,color="lightgray")
    ax.text(0.1,6371-660,"660 km",transform=trans,fontweight="bold",verticalalignment='center')
    ax.plot([0,1],[6371-2890,6371-2890],transform=trans,linewidth=12,color="lightgray")
    ax.text(0.1,6371-2890,"CMB",transform=trans,fontweight="bold",verticalalignment='center')
    ax.plot([0,1],[6371-5150,6371-5150],transform=trans,linewidth=6,color="lightgray")
    ax.text(0.1,6371-5150,"ICB",transform=trans,fontweight="bold",verticalalignment='center')   
```

### Density, gravity, pressure

```python
fig,axs = plt.subplots(1,3,figsize=(12,6))
axs[0].set_xlim(0,14)
axs[0].set_xlabel('Density $\\rho$ [10$^3$ kg/m$^3$]')
axs[0].set_ylim(0,6371)
axs[0].set_ylabel('Radius $r$ [m]')
boundaries(axs[0])
axs[0].plot(density/1000,radius,linewidth=4,color="red",label="$\\rho$")
axs[0].legend()

axs[1].set_xlim(0,13)
axs[1].set_xlabel('Gravity $g$ [m/s$^2$]')
axs[1].set_ylim(0,6371)
#axs[1].set_ylabel('Radius $r$ [m]')
boundaries(axs[1])
axs[1].plot(gravity,radius,linewidth=4,color="blue",label="$g$")
axs[1].legend()

axs[2].set_xlim(0,380)
axs[2].set_xlabel('Pressure $p$ [GPa]')
axs[2].set_ylim(0,6371)
#axs[2].set_ylabel('Radius $r$ [m]')
boundaries(axs[2])
axs[2].plot(pressure*1.e-9,radius,linewidth=4,color="green",label="$p$")
axs[2].legend()

plt.tight_layout()
```

### p-wave and s-wave velocities

```python
fig,axs = plt.subplots(1,3,figsize=(12,6))
axs[0].set_xlim(0,14)
axs[0].set_xlabel('p-wave velocity $v_p$ [km/s]')
axs[0].set_ylim(0,6371)
axs[0].set_ylabel('Radius $r$ [m]')
boundaries(axs[0])
axs[0].plot(vp/1000,radius,linewidth=4,color="red",label="$v_p$")
axs[0].legend()

axs[1].set_xlim(0,13)
axs[1].set_xlabel('s-wave velocity $v_s$ [km/s]')
axs[1].set_ylim(0,6371)
#axs[1].set_ylabel('Radius $r$ [m]')
boundaries(axs[1])
axs[1].plot(vs/1000,radius,linewidth=4,color="blue",label="$v_s$")
axs[1].legend()

plt.tight_layout()
```

### Bulk and shear modulus, Poisson ratio

```python
fig,axs = plt.subplots(1,3,figsize=(12,6))
axs[0].set_xlim(0,1500)
axs[0].set_xlabel('Bulk modulus $\\kappa$ [GPa]')
axs[0].set_ylim(0,6371)
axs[0].set_ylabel('Radius $r$ [m]')
boundaries(axs[0])
axs[0].plot(bulk/1e9,radius,linewidth=4,color="red",label="$\\kappa$")
axs[0].legend()

axs[1].set_xlim(0,300)
axs[1].set_xlabel('Shear modulus $\\mu$ [GPa]')
axs[1].set_ylim(0,6371)
#axs[1].set_ylabel('Radius $r$ [m]')
boundaries(axs[1])
axs[1].plot(shear/1e9,radius,linewidth=4,color="blue",label="$\\mu$")
axs[1].legend()

axs[2].set_xlim(0,0.55)
axs[2].set_xlabel('Poisson ratio $\\nu$ [-]')
axs[2].set_ylim(0,6371)
#axs[2].set_ylabel('Radius $r$ [m]')
boundaries(axs[2])
axs[2].plot(poisson,radius,linewidth=4,color="green",label="$\\nu$")
axs[2].legend()

plt.tight_layout()
```

### Temperatures

Here, we use our **geotherms**:

- continental geotherm: $T_{cont}$
- oceanic geotherm: $T_{ocean}$
- mantle adiabat: $T_{mantle}$

```python
# general
Kt = 3       # W/m/K
T0 = 273.    # K
alpha = 3e-5 # 1/K
cp = 1250.   # J/kg/K
rho = 3000.  # kg/m3
kappat = Kt/(rho*cp) # m2/s
print(kappat)

# continental geotherm
Q0 = 60      # mW/m2
H0 = 2.5e-6  # W/m3
d  = 10      # km


# oceanic geotherm
TA = 1573.   # K

# mantle geotherm
T1 = 1573.
g = 9.81
adia = T1*alpha*g/cp
print(adia*1.e3)

# phase boundaries
vel = 0.01                # m/yr
latent440 = +1.3e5        # J/kg
latent660 = -1.3e5        # J/kg

# rescale to SI units
Q0  = Q0/1000
d   = d*1000
vel = vel / (365.*24.*3600.)
```

```python
tempc = np.zeros(len(radius))
tempo = np.zeros(len(radius))
for i in range(len(radius)-2,-1,-1):
    if (radius[i] > 6251):
        depth = 6371 - radius[i]
        tempc[i] = T0 + H0*d**2/Kt * (1.-np.exp(-depth*1000/d)) + (Q0-H0*d)/Kt*depth*1000
        toffset  = tempc[i]
    elif (radius[i] <= 6251 and radius[i] > 3480):
        tempc[i] = toffset + adia*(6251-radius[i])*1000
    if (radius[i] < 6371-440):
        tempc[i] = tempc[i] + latent440 / cp
    if (radius[i] < 6371-660):
        tempc[i] = tempc[i] + latent660 / cp
    if (radius[i] < 6371-2890):
        tempc[i] = 3500 + (5700-3500)*(6371-2890-radius[i])/(6371-2890)

time40 = 40e6 * 365.*24.*3600.
for i in range(len(radius)-2,-1,-1):
    if (radius[i] > 6251):
        depth = 6371 - radius[i]
        tempo[i] = T0 + TA * scipy.special.erf((6371-radius[i])*1000/(2.*np.sqrt(kappat*time40)))
        toffset  = tempo[i]
    elif (radius[i] <= 6251 and radius[i] > 3480):
        tempo[i] = toffset + adia*(6251-radius[i])*1000
    if (radius[i] < 6371-440):
        tempo[i] = tempo[i] + latent440 / cp
    if (radius[i] < 6371-660):
        tempo[i] = tempo[i] + latent660 / cp
    if (radius[i] < 6371-2890):
        tempo[i] = 3500 + (5700-3500)*(6371-2890-radius[i])/(6371-2890)
```

```python
fig,axs = plt.subplots(1,3,figsize=(12,6))
axs[0].set_xlim(0,6000)
axs[0].set_xlabel('Temperature $T_c$ [K]')
axs[0].set_ylim(0,6371)
axs[0].set_ylabel('Radius $r$ [m]')
boundaries(axs[0])
axs[0].plot(tempc,radius,linewidth=4,color="red",label="$T_{cont}$")
axs[0].legend()

axs[1].set_xlim(0,6000)
axs[1].set_xlabel('Temperature $T_o$ [K]')
axs[1].set_ylim(0,6371)
#axs[1].set_ylabel('Radius $r$ [m]')
boundaries(axs[1])
axs[1].plot(tempo,radius,linewidth=4,color="blue",label="$T_{ocean}$")
axs[1].legend()

axs[2].set_xlim(0,0.55)
axs[2].set_xlabel('Poisson ratio $\\nu$ [-]')
axs[2].set_ylim(0,6371)
#axs[2].set_ylabel('Radius $r$ [m]')
boundaries(axs[2])
#axs[2].plot(poisson,radius,linewidth=4,color="green",label="$\\nu$")
axs[2].legend()

plt.tight_layout()
```

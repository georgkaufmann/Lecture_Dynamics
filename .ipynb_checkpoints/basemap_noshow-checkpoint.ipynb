```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import numpy as np
```

```python
# lon_0 is central longitude of projection.
# resolution = 'c' means use crude resolution coastlines.
plt.figure(figsize=(12,6))
map = Basemap(projection='robin',lon_0=0,resolution='c')
map.drawcoastlines()
map.fillcontinents(color='coral',lake_color='blue')
# draw parallels and meridians.
map.drawparallels(np.arange(-90.,120.,30.))
map.drawmeridians(np.arange(0.,360.,60.))
map.drawmapboundary(fill_color='aqua')
plt.title("Robinson Projection")
plt.show()
```

```python
# setup Lambert Conformal basemap.
plt.figure(figsize=(12,6))
map = Basemap(width=12000000,height=9000000,projection='lcc',
            resolution='c',lat_1=45.,lat_2=55,lat_0=50,lon_0=-107.)
# draw coastlines.
map.drawcoastlines()
# draw a boundary around the map, fill the background.
# this background will end up being the ocean color, since
# the continents will be drawn on top.
map.drawmapboundary(fill_color='aqua')
# fill continents, set lake color same as ocean color.
map.fillcontinents(color='coral',lake_color='aqua')
plt.show()
```

```python
# setup Lambert Conformal basemap.
# set resolution=None to skip processing of boundary datasets.
plt.figure(figsize=(12,6))
map = Basemap(width=12000000,height=9000000,projection='lcc',
            resolution=None,lat_1=45.,lat_2=55,lat_0=50,lon_0=-107.)
map.shadedrelief()
plt.show()
```

```python
# setup Lambert Conformal basemap.
# set resolution=None to skip processing of boundary datasets.
plt.figure(figsize=(12,6))
map = Basemap(width=12000000,height=9000000,projection='lcc',
            resolution=None,lat_1=45.,lat_2=55,lat_0=50,lon_0=-107.)
map.etopo()
plt.show()
```

```python
plt.figure(figsize=(14,6))
map = Basemap(projection='robin',lon_0=180,resolution='c')
map.drawcoastlines()

nlats = 5; nlons = 5; delta = 2.*np.pi/(nlons-1)
lats = np.linspace(-80,80,11)
lons = np.linspace(1,359,11)
lons, lats = np.meshgrid(lons, lats)

# compute native map projection coordinates of lat/lon grid.
x, y = map(lons, lats)

map.scatter(x,y,s=200,latlon=False,alpha=0.5)
```
